## 우주 정거장에 있는 우주인 확인하기

```python
import requests # requests 웹데이터를 가져올수있는 패키지 
import pandas as pd

# Make a get request to get the latest position of the international space station from the opennotify api.
response = requests.get("http://api.open-notify.org/astros.json")
```

```python
response
```

```python
# Get the response data as a python object.  Verify that it's a dictionary.
info= response.json() # json 은 반정형 
info

```

```python
#using pandas!!!
people=info['people']
print(people)
```

```python
import pandas as pd # 판다스 패키지를 pd라는 별칭으로 부르겠다. 

df=pd.DataFrame(people) # pd에서 DataFrame함수를 사용하겠다. 
print(f"There are {info['number']} people in the space right now ")

display(df.reindex(index=None,columns=['name','craft'])) #convert oder of columns and index
```
