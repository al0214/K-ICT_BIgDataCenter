````python
# from sympy라는 패키지에서
# 다음과 같은 함수를 실행 할꺼야라고 선언
from sympy import Symbol, symbols, Eq, solve, discriminant,solveset
````

````python
x, y = symbols('x y')
````

$$2x + y - 1 = 0$$

````python
eq1 = Eq(2*x - y - 1,0) 
````

$$x + y - 5=0$$

````python
eq2 = Eq(x + y - 5,0)
````

````python
sol = solve((eq1, eq2),(x, y))
sol
# solve 형태는 딕셔너리이다
````

````python
print(f'The solution is x = {sol[x]}, y = {sol[y]}')
````

$$x^2+7x+6=0$$

````python
x= Symbol('x')


roots= solve(x**2+7*x+6,(x))
roots
````

$$2x^2 + x + 1=0$$

````python
x = Symbol("x")
expression = 2 * x ** 2 + x + 1
````

$$D= b^2 - 4ac$$

- $D>0$: 서로다른 두실근 
- $D=0$: 중근
- $D<0$: 허근 

````python
discriminant(expression)
````

````python
answer = solve(expression,(x))
````

````python
answer
````

````python
solveset(expression, x)
````
